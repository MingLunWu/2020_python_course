## 更換DataFrame欄位名稱

值得注意的是：欄位名稱如果使用中文，容易發生意料之外的錯誤（尤其是在Query時)

```python
df.rename(columns={"舊欄位名稱":"新欄位名稱", "舊欄位名稱2":"新欄位名稱2"})
```

使用`df.columns`可以得到所有的欄位名稱：

````python
df.columns
````

使用 `df.rename()` 將欄位名稱更換掉:

````python
df = df[["發票號碼", "日期", "公司名稱", "銷售金額", "營業稅"]] # 只取特定的幾個欄位
df
````
